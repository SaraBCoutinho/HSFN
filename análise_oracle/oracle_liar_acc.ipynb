```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_name = 'liar'

def load_predictions(dataset_name):
    #path =  '/content/Data/liar/' +'pred_val_2_a.csv'
    path =  'pred_val_a.csv'
    table_pred = pd.read_csv(path)
    label = table_pred['label']
    methods = table_pred.drop('label', axis=1)
    methods = methods.drop('Unnamed: 0', axis=1)
    return label, methods

def get_oracle(labels, pred={}, clfs=[]):
    cont = 0;
    list_yes = np.zeros((len(labels)))
    oracle_preds = np.zeros((len(labels)))
    for i, k in enumerate(labels):
        acertou = False
        for clf in clfs:
            if pred[clf][i] == k:
                acertou = True
                oracle_preds[i] = pred[clf][i]
                break
        if acertou:
            cont += 1
            list_yes[i] = k
        else:
            oracle_preds[i] = pred[clf][0]
            if k == 0:
                list_yes[i] = 1
            else:
                list_yes[i] = 0
    return cont, list_yes, oracle_preds

clfs = ['CV-SVM', 'TFIDF-SVM', 'W2V-SVM', 'GLOVE-SVM', 'FAST-SVM',
        'CV-KNN', 'TFIDF-KNN','W2V-KNN', 'GLOVE-KNN', 'FAST-KNN',
        'CV-MLP','TFIDF-MLP','W2V-MLP','GLOVE-MLP','FAST-MLP',
        'CV-NB','TFIDF-NB','W2V-NB','GLOVE-NB','FAST-NB',
        'CV-EXTRA','TFIDF-EXTRA','W2V-EXTRA','GLOVE-EXTRA','FAST-EXTRA',
        'CV-RF','TFIDF-RF','W2V-RF','GLOVE-RF','FAST-RF',
        'CV-LR','TFIDF-LR','W2V-LR','GLOVE-LR','FAST-LR',
        'CV-CNN','TFIDF-CNN','W2V-CNN','GLOVE-CNN','FAST-CNN']
```
