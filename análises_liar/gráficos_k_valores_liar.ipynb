```python
# importao de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# arquivo 
df = pd.read_csv('Liar_k_valores.csv') #pd.read_csv(/content/Liar_k_values.csv')

# formatando os valores: [0,1]
df['Accuracy(LR)'] = df['Accuracy(LR)']/1000
df['Accuracy(RF)'] = df['Accuracy(RF)']/1000
df['Accuracy(NB)'] = df['Accuracy(NB)']/1000
df['Precision(LR)'] = df['Precision(LR)']/1000
df['Precision(RF)'] = df['Precision(RF)']/1000
df['Precision(NB)'] = df['Precision(NB)']/1000
df['Recall(LR)'] = df['Recall(LR)']/1000
df['Recall(RF)'] = df['Recall(RF)']/1000
df['Recall(NB)'] = df['Recall(NB)']/1000
df['F1-score(LR)'] = df['F1-score(LR)']/1000
df['F1-score(RF)'] = df['F1-score(RF)']/1000
df['F1-score(NB)'] = df['F1-score(NB)']/1000

# grficos
fig, ax = plt.subplots(nrows=1, ncols=4, squeeze = False, figsize=(20, 4))
#---df
ax[0][0].plot(df['k'],df['Accuracy(LR)'],'r',label='LR')
ax[0][0].plot(df['k'],df['Accuracy(RF)'],'b',label='RF')
ax[0][0].plot(df['k'],df['Accuracy(NB)'],'g',label='NB')
ax[0][0].set_xlabel('k', color = 'black')
ax[0][0].set_ylabel('Liar', color = 'black')
ax[0][0].legend(loc='best')

ax[0][1].plot(df['k'],df['Precision(LR)'],'r',label='LR')
ax[0][1].plot(df['k'],df['Precision(RF)'],'b',label='RF')
ax[0][1].plot(df['k'],df['Precision(NB)'],'g',label='NB')
ax[0][1].set_xlabel('k', color = 'black')

ax[0][2].plot(df['k'],df['Recall(LR)'],'r',label='LR')
ax[0][2].plot(df['k'],df['Recall(RF)'],'b',label='RF')
ax[0][2].plot(df['k'],df['Recall(NB)'],'g',label='NB')
ax[0][2].set_xlabel('k', color = 'black')

ax[0][3].plot(df['k'],df['F1-score(LR)'],'r',label='LR')
ax[0][3].plot(df['k'],df['F1-score(RF)'],'b',label='RF')
ax[0][3].plot(df['k'],df['F1-score(NB)'],'g',label='NB')
ax[0][3].set_xlabel('k', color = 'black')

ax[0][0].set_title("Acurcia x k ") 
ax[0][1].set_title("Preciso x k")
ax[0][2].set_title("Revocao x k")
ax[0][3].set_title("Medida F1 x k")

plt.savefig('grficos_k_valores_liar.pdf')
plt.show()
```
